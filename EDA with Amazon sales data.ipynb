# INTERPRET THE PLOT

- The plot shows a bimodal distribution, indicating two peaks in the amount values

- This suggests the presence of two different subpopulations within the data

- The wider sections of the plot indicate higher density (more data points)

- Narrower sections indicate lower density

- The peaks suggest where the majority of the data is concentrated

- The long tail extending toward the higher values indicates the presence of outliers

- The central line within the violin indicates the median of the data

# INSIGHTS

- There are two predominant groups of amounts, which may indicate differing behaviors or categories in your dataset

- The long tail suggests that while the majority of values are clustered around the lower amounts,there are significant instances of very high amounts

# TYPE OF MISSING VALUE AND ITS SOLUTION

- Given the amount of missing values (v.high) and the distribution characteristics,the missing values in the Amount column are likely MCAR (Missing Completely at Random):

- The randomness in missing values might not depend on the observed data, making it independent of the data distribution

# Solution: handled by simple IMPUTATION

- A common method is to fill in the missing values with the mean or median of the non-missing values

```python
df.fillna({'Amount': df['Amount'].mean()}, inplace=True)
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# I.) Univariate analysis

```python
# Q1.) What is the distribution of order amounts?

# A.)  Histogram + Boxplot for outliers

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['Amount'], bins=30, kde=True)
plt.title('Amount Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Amount'])
plt.title('Amount Outliers')

plt.show()
```

# Output interpretation 

- distribution is right-skewed, indicating that most orders are concentrated at lower amounts (under 1000) with fewer high-value orders

- significant number of outliers, suggesting extreme values

```python
# Q2.) Which product categories are most popular?

# A.) Horizontal bar plot

plt.figure(figsize=(8, 4))

df['Category'].value_counts().plot(kind='barh')
plt.title('Top Product Categories')

plt.show()
```

# Output interpretation

- There is a strong preference for casual wear, particularly T-shirts, while accessories and formal wear have lower popularity

# II.) Bivariate Analysis

```python
# Q3.) Is there a relationship between order amount and product category?

# A.) Boxplot by category

plt.figure(figsize=(10, 5))

sns.boxplot(data=df, x='Category', y='Amount')
plt.xticks(rotation=45)
plt.title('Amount by Product Category')

plt.show()
```

# Output interpretation

- There are numerous outliers across categories, prominently in T-shirts and Blazers, indicating certain transactions that significantly exceed typical amounts.

- Perfume and Wallet categories exhibit tighter distributions, suggesting more consistent pricing and less variability in customer spending compared to other categories.

```python
df['B2B'].value_counts()
```

```python
# Q4.) Are B2B orders more expensive than B2C?

# A.) Grouped Bar Plot

sns.barplot(data=df, x='B2B', y='Amount', estimator=sum)
plt.title('Total Revenue: B2B or Not B2B')
plt.show()
```

# Output interpretation

- The revenue generated from non-B2B sources (False) is significantly higher than that from B2B sources (True).
- This indicates that the business likely relies much more on non-B2B revenues.
- This could be due to B2B orders being more expensive.

```python
# Q5.) Which states have the highest order volumes?

# A.) Top 10 staes bar plot

plt.figure(figsize=(10, 4))

df['Ship-State'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 States by Order Volume')

plt.show()
```

# Output interpretation

- Maharashtra has the highest order volume, significantly surpassing other states, with over 20,000 units.
- Karnataka follows, with a volume around 15,000 units.
- Tamil Nadu, Telangana, and Uttar Pradesh display moderate order volumes ranging from approximately 10,000 to 12,000 units.
- Delhi, Kerala, West Bengal, and Andhra Pradesh showcase lower volumes, each around 5,000 to 10,000 units.
- Gujarat has the least order volume among the top 10 states.

# III.) Multivariate Analysis

```python
# Q6.) Is there a correlation between order amount, quantity, and B2B status?

# A.) Pairplot with hue

plt.figure(figsize=(12, 8))
sns.pairplot(df[['Amount', 'Qty', 'B2B']], hue='B2B')
plt.suptitle('Relationships: Amount, Quantity, and B2B Status', va = 'top')
plt.subplots_adjust(top=0.9)

plt.show()
```

# Output interpretation

- Amount: Highly skewed distribution (many values near 0).
- Quantity: Also clustered towards lower values.
- Low correlation observed between Amount and Quantity.
- The orange dot (B2B True) is an outlier with a higher Amount but low Quantity.
